```python
import pandas as pd

url = "https://drive.google.com/uc?id=1-crPzL6qMinByPzsrEHhGn1EJ1MfD3GX"
df = pd.read_csv(url, names=list(range(0, 100, 1)))
city_map_list = df.values.tolist() # двумерный список (матрица) 0 означает реку, 1 - землю

courier_location = (0, 17) # стартовая позиция курьера
orders_location = [(86, 18)] # список координат для доставки товаров

def get_valid_moves(now_pos, city_map_list):
    """Определяем доступные соседние клетки для следующего шага"""
    
    city_size_x = len(city_map_list[0]) # размер города по оси х
    city_size_y = len(city_map_list) # размер города по оси y
    
    curr_x, curr_y = now_pos  
    deltas = [(1, 0), (0, 1), (-1, 0), (0, -1)]
    valid_moves = []
    
    for dx, dy in deltas:
        nx, ny = curr_x + dx, curr_y + dy
        if 0 <= nx < city_size_x and 0 <= ny < city_size_y: # проверка, что новые точко находятся в границах города
            if city_map_list[ny][nx] == 1: # проверка проходимости (0 или 1 на карте города)
                valid_moves.append((nx, ny))
                
    return valid_moves

def navigator(courier_location, orders_location):
    """Последовательно находит кратчайшие пути до всех точек заказов"""
    
    temp_paths = [[courier_location]] # список всех возможных путей путей (пока не дошли до точки доставки)
    final_paths = []  # итоговый список путей 
    visited = {courier_location} # множество посещенных клеток
    
    while len(orders_location) > 0: # цикл работает пока все заказы не будут доставлены
        order_found = False
        temp_route = [] # временный маршрут
    
        for path in temp_paths:                                 # проходим по всем возможным путям
            next_pos = get_valid_moves(path[-1], city_map_list) # для последней точки каждого пути находим доступные клетки
            for pos in next_pos:                                # проходим по всем доступным клеткам
                if pos in visited:                              # если эту клетку уже посещали, переходим к следующей возможной клетке
                    continue
                    
                new_path = path + [pos]               # добавляя текущую клетку к старому пути
                
                if pos in orders_location:            # если текущая клетка в списке заказов
                    orders_location.remove(pos)       # удаляем из списка заказов координату доставленного заказа
                    final_paths.append(new_path[1:])  # сохраняем путь доставки(не включаем начальную точку)
                    temp_route = [[pos]]              # сбрасываем временный маршрут, оставляем последнюю точку доставки
                    visited = {pos}                   # сбрасываем множество посещенных клеток, оставляем последнюю точку доставки
                    order_found = True                # флаг доставки
                    break                             # выходим из цикла проверки доступных клеткок
                else:                                 # сли текущей клетки нет в списке заказов 
                    temp_route.append(new_path)       # добавляем последнюю точку во временный маршрут
                    visited.add(pos)                  # отмечаем клетку как посещенную
    
            if order_found:
                break # выходим из цикла проверки возможных путей
    
        temp_paths = temp_route # обновляем список всех возможных путей
    return final_paths
        
print("Финальный маршрут:")
route = []
for sub in navigator(courier_location, orders_location):
    route.extend(sub) # соединяем все найденые пути в один маршрут
print(*route)
```
